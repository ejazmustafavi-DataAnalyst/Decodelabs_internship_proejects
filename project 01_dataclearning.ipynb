# Data clearning & preparation for Orders dataset

```python
import pandas as pd
import numpy as np
```

```python
# import data 
df=pd.read_excel(r'E:\DATA Science\portfolio project\decodelab\ordersdatas.xlsx')
print("Shape of the dataset",df.shape)
print(df.head(5))
```

```python
# datatypes
df.dtypes
```

```python
# checking null values
df.isnull().sum()
```

```python
# filling CouponCode null values with "NO Coupan"
df['CouponCode']=df['CouponCode'].fillna('NO Coupan')

#Recheking missing values
df.isnull().sum()
```

```python
# checking duplicates
df.duplicated().sum()
```

```python
# Check unique values in text columns
print(df['OrderStatus'].unique())
print(df['PaymentMethod'].unique())
print(df['ReferralSource'].unique())
print(df['Product'].unique())
```

```python
# checking outliers in Quantity, UnitPrice, and TotalPrice

df[['Quantity', 'UnitPrice', 'TotalPrice']].describe()
```

```python
# validating TotalPrice matches Quantity × UnitPrice

df['calculated'] = df['Quantity'] * df['UnitPrice']
mismatch = df[df['calculated'].round(2) != df['TotalPrice'].round(2)]
print(mismatch[['OrderID','Quantity','UnitPrice','TotalPrice','calculated']])
```

```python
# remove calculated column
df.drop(columns=['calculated'], inplace=True)
```

```python
# Extract date features from Date column

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.day_name()
df['WeekNum'] = df['Date'].dt.isocalendar().week
```

```python
# Create a Revenue bin / category for TotalPrice

df['PriceCategory'] = pd.cut(df['TotalPrice'], 
                              bins=[0, 500, 1500, 3500],
                              labels=['Low', 'Medium', 'High'])
```

```python
print("Total Revenue:", df['TotalPrice'].sum())
print("Total Orders:", df['OrderID'].nunique())
print("Average Order Value:", df['TotalPrice'].mean())
print("Most Sold Product:", df['Product'].value_counts().idxmax())
```
